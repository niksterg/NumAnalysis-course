### Exercise 1

Find the roots of the equation $e^{2x}-3x-1=0$ with the fixed-point iteration method.

```python
def f1(x):
    return np.exp(2*x) - 3.0*x-1
```

```python
xpoints = np.linspace(-0.5,0.5,500) 
fpoints = f1(xpoints)

plot1 = plt.plot(xpoints, fpoints)
plt.grid(True)
plt.show(plot1)
```

#### First root:

```python
iteration = [0]
x = 0.1
solution = [x]

for i in range(0,60):
   x = (np.exp(2.0*x)-1.0)/3.0
   iteration.append(i+1)
   solution.append(x)
   print('%.d %.16f'% (iteration[i], solution[i]))
    
```

```python
N = len(solution)
error = solution[N-1]-solution

fig, ax = plt.subplots()
ax.plot(iteration, abs(error))
ax.set_yscale('log')
ax.set(xlabel='iteration', ylabel='error')
ax.grid()
plt.ylim(bottom=abs(error[-2]))
plt.show()
```

```python
display(N)
```

#### Confirm that error is $\epsilon_{n+1} = g'(\rho)\epsilon_{n}$

```python
root = solution[N-1]
dgdx = 2.0*np.exp(2.0*root)/3.0

N = len(solution)
error_n = np.array([solution[N-1]-solution[0]])
error_np1 = np.array([solution[N-1]-solution[1]])

for i in range (1,N-1):
    error_n = np.append(error_n, [solution[N-1]-solution[i]])
    error_np1 = np.append(error_np1, [solution[N-1]-solution[i+1]])

error_np1_predict = dgdx*error_n

fig, ax = plt.subplots()
ax.plot(abs(error_n), abs(error_np1_predict))
ax.plot(abs(error_n),abs(error_np1))
ax.set_yscale('log')
ax.set_xscale('log')
ax.set(xlabel='error_n', ylabel='error_n+1')
plt.xlim(left=abs(error_n[-2]))
ax.grid()
plt.show()
```

## Exercise 2 
Find the roots of the equation $e^{2x}-3x-1=0$ with Newton-Raphson's method.

#### First root:

```python
iteration = [0]
x=0.1
solution = [x]
print('%.d %.16f'% (iteration[0], solution[0]))
for i in range(1,6):
   x = x- (np.exp(2.0*x)-3.0*x-1.0)/(2.0*np.exp(2.0*x)-3.0)
   iteration.append(i)
   solution.append(x)
   print('%.d %.16f'% (iteration[i], solution[i]))
```

```python
N = len(solution)
error = solution[N-1]-solution
fig, ax = plt.subplots()
ax.plot(iteration, abs(error))
ax.set_yscale('log')
ax.set(xlabel='iteration', ylabel='error')
ax.grid()
plt.xlim(right=N-2)
tick_spacing = 1
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
plt.show()
```

#### Confirm that error is $\epsilon_{n+1} = -\frac{f''(\rho)}{2f'(\rho)}\epsilon^2_{n}$

```python
root = solution[N-1]
k = -4.0*np.exp(2.0*root)/(2.0*(2.0*np.exp(2.0*root)-3.0))
                           
N = len(solution)
error_n = np.array([solution[N-1]-solution[1]])
error_np1 = np.array([solution[N-1]-solution[2]])
for i in range (2,N-1):
    error_n = np.append(error_n, [solution[N-1]-solution[i]])
    error_np1 = np.append(error_np1, [solution[N-1]-solution[i+1]])

error_np1_predict = k*error_n**2

NE = len(error_n)
fig, ax = plt.subplots()
ax.plot(error_n, error_np1_predict)
ax.plot(error_n, error_np1)
ax.set(xlabel='error_n', ylabel='error_n+1')
ax.set_xscale('log')
ax.set_yscale('log')
plt.ylim(bottom=error_np1[NE-2])
plt.xlim(left=error_np1[NE-3])
ax.grid()
plt.show()
```
